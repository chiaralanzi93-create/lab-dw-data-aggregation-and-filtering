```python
# Step 1: Count unique customers per state
customer_counts = df_original.groupby('State')['Customer'].nunique().reset_index(name='Total Customers')

# Step 2: Filter states with more than 500 customers
filtered_states = customer_counts[customer_counts['Total Customers'] > 500]

# Step 3: Optionally, you can sort the results for better readability
filtered_states_sorted = filtered_states.sort_values(by='Total Customers', ascending=False)

# Display the result
print(filtered_states_sorted)
```

```python
#4 


# Group by 'Gender' and 'Education', then calculate the required statistics
statistics = df_original.groupby(['Gender', 'Education'])['Customer Lifetime Value'].agg(['max', 'min', 'median']).round(2)

statistics_in_thousands = statistics / 1000
statistics_in_thousands = statistics_in_thousands.round(3)
# Display the results
print(statistics_in_thousands)
```

```python
# 5 Bonus The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

summary_marketing = df_original.pivot_table(index = 'State', columns = 'Months Since Policy Inception', values ='Policy',aggfunc = 'count')
summary_marketing

```

```python
# 6 To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.


summary_new = df_original.groupby(['State','Months Since Policy Inception'])['Policy'].count().reset_index(name = 'count_policy')

summary_new




```

```python
# alternative solution 

summary_new = df_original.groupby(['State','Months Since Policy Inception'])['Policy'].count().reset_index(name = 'count_policy')

state_totals = summary_new.groupby('State')['count_policy'].sum()

top_states = state_totals.nlargest(3).index

filtered_summary = summary_new[summary_new ['State'].isin(top_states)]

filtered_summary = filtered_summary.sort_values(by = 'count_policy', ascending = False)

print(filtered_summary)
```

```python
import pandas as pd

# Group by 'State' and 'Months Since Policy Inception' and count policies
summary_new = df_original.groupby(['State', 'Months Since Policy Inception'])['Policy'].count().reset_index(name='Total_Policies')

# Calculate total policies per state
state_totals = summary_new.groupby('State')['Total_Policies'].sum()

# Find the top 3 states with the highest number of policies
top_states = state_totals.nlargest(3).index

# Filter the original summary to include only the top 3 states
filtered_summary_new = summary_new[summary_new['State'].isin(top_states)]

# Sort the resulting DataFrame for better readability
filtered_summary_new = filtered_summary_new.sort_values(by=['State', 'Months Since Policy Inception'])

# Display the final DataFrame
print(filtered_summary_new)
```

```python
#7 

#Filter for Positive Responses: Focus on only those customers who responded "Yes".
import pandas as pd

# Filter the DataFrame to include only positive responses
positive_response = df_original[df_original['Response'] == 'Yes']

# Count the positive responses per coverage type (assuming 'Coverage' is intended here)
marketing_table = positive_response.groupby(['Coverage'])['Response'].count().reset_index(name='Count_response')

# Count the total responses per coverage type for response rate calculation
total_responses = df_original.groupby(['Coverage'])['Response'].count().reset_index(name='Total_Response_Count')

# Merge the dataframes on Coverage to bring them into one table and calculate the rate%
response_rate_data = pd.merge(marketing_table, total_responses, on='Coverage')

# Calculate the response rate as a percentage
response_rate_data['Response_Rate'] = ((response_rate_data['Count_response'] / response_rate_data['Total_Response_Count']) * 100).round(2)

# Sort the final table by Response Rate
response_rate_table = response_rate_data[['Coverage', 'Response_Rate']].sort_values(by='Response_Rate', ascending=False)

# Display the table
print(response_rate_table)
```
